```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = np.loadtxt('binarydigits.txt') # Load data
```

```python
def learn_pML(data): # Function to learn ML probability
    N, D = data.shape
    pML = (1/N)*np.sum(data, axis=0)
    return pML

def learn_pMAP(data, alpha, beta): # Function to learn MAP probability
    N, D = data.shape
    pML = learn_pML(data)
    pMAP = (N*pML + alpha - 1)/(N + alpha + beta - 2)
    return pMAP

def plot_pixels(data, title=''): # Function to plot results
    plt.figure()
    plt.imshow(np.reshape(data, (8,8)), vmin=0, vmax=1, interpolation='None', cmap='grey')
    plt.title(title)
```

```python
pML = learn_pML(data) 
plot_pixels(pML, "Maximum Likelikehood Probability Pixel Colours")

pMAP = learn_pMAP(data, 3.0, 3.0)
plot_pixels(pMAP, "Maximum A Posteriori Probability Pixel Colours")
```
